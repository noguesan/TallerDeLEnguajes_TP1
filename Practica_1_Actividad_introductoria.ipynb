2- Haz un programa que pida al usuario que ingrese una temperatura en
grados Celsius y luego convierta esa temperatura a grados Fahrenheit,
mostrando el resultado.

```python
celsius = input("Ingresa la temperatura en grados Celsius: ")
celsius = float(celsius)
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C equivale a {fahrenheit:.2f}°F")          # :.2f pra que el numero tenga 2 decimales
```

5- Implementa un programa que solicite al usuario que ingrese una lista de
números. Luego, imprime la lista pero detén la impresión si encuentras un
número negativo. Nota: utilice la sentencia break cuando haga falta.

```python
# Solicitar una lista de números al usuario
numeros = list(map(int, input("Ingresa números separados por espacio: ").split()))

"""
Función split()
split() es un método de cadenas que divide una cadena en partes basadas en un separador.
Por defecto, split() usa los espacios como separador.
Ejemplo
 entrada = "10 20 30 40"
 lista = entrada.split()
 print(lista)  # ['10', '20', '30', '40']

Función map(función, iterable)
La función map() aplica una función a cada elemento de un iterable (como una lista o una tupla)
 y devuelve un objeto iterable con los resultados.
En este caso, map(int, ...) convierte cada elemento de la lista de cadenas en un número entero.
Ejemplo
  lista_de_cadenas = ['10', '20', '30', '40']
  lista_de_enteros = list(map(int, lista_de_cadenas))
  print(lista_de_enteros)  # [10, 20, 30, 40]

"""

# Recorrer la lista
for num in numeros:
    if num < 0:  # Si el número es negativo
        break  # Detiene el bucle
    print(num)  # Imprime solo si es positivo
```

6- Modifique el ejercicio 4 para que dada la lista de número genere dos nuevas
listas, una con los número pares y otras con los que son impares. Imprima
las listas al terminar de procesarlas.

```python
# Lista de números de ejemplo
lista_numeros = [10, 15, 22, 33, 40, 55, 60, 75, 80]

# Inicializar listas para pares e impares
pares = []
impares = []

# Recorrer la lista
for num in lista_numeros:
    if num % 2 == 0:  # Si el número es par
        pares.append(num)
    else:
        impares.append(num)

# Imprimir las listas
print("Números pares:", pares)
print("Números impares:", impares)
```
